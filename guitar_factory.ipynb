```python
import matplotlib.pyplot as plt
import seaborn as sns
from constants import *
from main import simulate
```

```python
s1 = simulate(default_constants)
```

```python
def plot_dict_days_info(s):
    for key in s['days_info_dict']:
        plt.figure(figsize=(6,4))
        ax = sns.barplot(y=s['days_info_dict'][key], x=[i + 1 for i in range(len(s['days_info_dict'][key]))])
        ax.set_xlabel("День")
        ax.set_ylabel(labels[key])
        plt.show()
        
def plot_overall_info(s, key):
    plt.figure(figsize=(6,4))
    keys = list(s[key].keys())
    keys_ru = list(map(lambda x: labels[x], keys))
    vals = [s[key][k] for k in keys]
    ax = sns.barplot(x=keys_ru, y=vals)
    for item in ax.get_xticklabels():
        item.set_rotation(45)
    ax.set_ylabel("Всего")
    plt.show()

def plot_all_diagrams(s):
    print('Диаграммы по дням')
    plot_dict_days_info(s)
    print('Диаграммы после завершения симуляции')
    plot_overall_info(s, 'in_factory')
    plot_overall_info(s, 'overall')
```

```python
plot_all_diagrams(s1)
```
